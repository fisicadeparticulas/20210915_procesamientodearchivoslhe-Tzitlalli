```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import plotly.express as px
 
data=pd.read_csv("https://goo.gl/g8y9TC")
 
data.head(100)
data.count()
```

```python
data_g = data.loc[data['type2'] != 'T']

data_g.head(100)
```

```python

E= (data_g['E1'] + data_g['E2'])**2
Px= (data_g['px1']+data_g['px2'])**2
Py= (data_g['py1']+data_g['py2'])**2
Pz= (data_g['pz1']+data_g['pz2'])**2
P = Px + Py + Pz
EP = E-P

m = np.sqrt(EP)

fig_1=px.histogram(m,m, nbins = 1000)
fig_1.show()
#data_g.count()

#plt.hist(m)
```

```python
fig_1=px.histogram(data_g['pt1'],data_g['pt1'],nbins=1000, range_x=["0","100"],labels={"x": "$P_{T1}$" }  )
fig_1.update_layout(title=r"$\text{Momento transverso} P_{T_1}$ , \ \text{%d bins}$" %100,xaxis_title=r"$\text{P} \ \text{(GeV)} $",yaxis_title="Eventos")
fig_1.show()

fig_11=px.histogram(data_g['pt2'],data_g['pt2'],nbins=10000, range_x=["0","100"],labels={"x": "$P_{T2}$" }  )
fig_11.update_layout(title=r"$\text{Momento transverso} P_{T_2}$ , \ \text{%d bins}$" %100,xaxis_title=r"$\text{p} \ \text{(GeV)} $",yaxis_title="Eventos")
fig_11.show()
```

```python
fig_1=px.histogram(data_g['phi1'], data_g['phi1'], nbins=100,  labels={"x": "radianes", } )
fig_1.update_layout(title=r"$\phi_1$")
fig_1.show()

fig_11=px.histogram(data_g['phi2'], data_g['phi2'], nbins=100,  labels={"x": "radianes", } )
fig_11.update_layout(title=r"$\phi_2$")
fig_11.show()

dphi = np.abs(data_g['phi2'] - data_g['phi1'])
fig_12=px.histogram(dphi, dphi, labels={"x": "radianes", })
fig_12.update_layout(title=r"$\delta \phi$")
fig_12.show()
```

```python
p=np.sqrt(data_g['px1']**2+data_g['py1']**2+data_g['pz1']**2)
pz=data_g['pz1']

x=pz/p
theta=np.arccos(x)
def eta(x):
  eta=-np.log(np.tan(np.arccos(x)/2))
  return eta

eta=pd.DataFrame(eta(x))

fig_0=px.scatter(x=theta,y=eta)
fig_0.update_layout(
    title=r"$\theta  \ vs \ \eta$",
    xaxis_title=r"$\theta \text{[rad]} $",
    yaxis_title=r"$ \eta $")
fig_0.show()
```

```python
bins=([14,30,100,1000])
for i in range(len(bins)):
  print(bins[i])
  fig_1=px.histogram(eta,x=eta[0],nbins=bins[i])
  fig_1.update_layout(title=r"$\text{DistribuciÃ³n de} \ \eta, \ \text{%d bins}$" %bins[i],
                      xaxis_title=r"$\eta$",
                      yaxis_title="Eventos")
  fig_1.show()
  
```
